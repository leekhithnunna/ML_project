```python
import pandas as pd
import fasttext
import numpy as np

df = pd.read_excel('/content/bert_features_output.xlsx')
df = df.head(5000)
# Load FastText model
ft_model = fasttext.load_model('/content/cc.en.300.bin')

# Convert each review to FastText vector
def get_ft_vector(text):
    return ft_model.get_sentence_vector(str(text))

# Apply to all rows
features = df['text_'].apply(get_ft_vector)
feature_matrix = np.vstack(features.values)

# Turn into DataFrame
feature_df = pd.DataFrame(feature_matrix, columns=[f'ft_{i}' for i in range(feature_matrix.shape[1])])

# Combine with original data
final_df = pd.concat([df[['text_', 'label']], feature_df], axis=1)

# Save to Excel
final_df.to_excel('fasttext_features_output.xlsx', index=False)
```
